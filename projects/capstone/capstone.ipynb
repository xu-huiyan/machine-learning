```python
import pandas as pd
from IPython.display import display
from scipy import stats
import numpy as np
import math
```

## Preparation

```python
data=pd.read_csv("train_v2.csv")
data = data.drop('id', axis = 1)
features=data.drop('loss',axis=1)
loss=data['loss']
```

```python
display(features.describe())
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, loss, test_size=0.2, random_state=0)
```

## Benchmark Model

```python
#Generate a probability function 

l=[]
p=[]
i_n=-0.1
for i in xrange(0,1001,1):
    l.append(i/10.0)
    p.append(int(sum(x > i_n and x <=i/10.0 for x in y_train)))
    i_n=i/10.0
```

```python
prob = [float(1.0*pr / len(y_train)) for pr in p]
mydis = stats.rv_discrete(name='myd', a=0,b=100,values=(l, prob))
```

```python
random_loss = mydis.rvs(size=len(y_test),random_state=0)
```

```python
def MAE(y_predicted, y_actual):
    return round(1.0*sum(abs(y_predicted-y_actual))/len(y_actual),9)

#default-nondefault accuracy
def dnd_accuracy(y_predicted, y_actual):
    predicted_dnd = np.array([math.ceil(y/100.0) for y in y_predicted])
    actual_dnd= np.array([math.ceil(y/100.0) for y in y_actual])
    return (1-round(1.0*sum(abs(predicted_dnd-actual_dnd))/len(actual_dnd),9))
```

```python
print(MAE(random_loss, y_test))
print(dnd_accuracy(random_loss, y_test))
```
